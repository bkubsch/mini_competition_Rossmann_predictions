```python
store = pd.read_csv(
    "data/store.csv"
)
new_train = pd.read_csv("data/new_train.csv",index_col=0, dtype={"StateHoliday":object})
data = store.merge(new_train,on="Store")

data.head()
```

```python
data["Cat_Comp_Month"] = np.where(data["CompetitionOpenSinceMonth"] >= 0, 1, 0)
```

```python
data.head()
```

```python
data["Cat_Comp_Year"] = np.where(data["CompetitionOpenSinceYear"] >= 0, 1, 0)
```

```python
data.head()
```

```python
comp_year_sales = data.groupby("CompetitionOpenSinceYear").mean()["Sales"].to_frame().reset_index()

sns.set(rc={'figure.figsize':(14.7,8.27)})
sns.barplot(comp_year_sales["CompetitionOpenSinceYear"], comp_year_sales["Sales"])
```

```python
categoricals_to_graph = ["Open", "DayOfWeek", "StateHoliday", "SchoolHoliday", 
            "Promo", "Promo2", "StoreType", "Assortment", "Cat_Comp_Year", "Cat_Comp_Month", 
                         "Promo2SinceWeek", "Promo2SinceYear"]

for column in categoricals_to_graph: 
    comp_year_sales = data.groupby(column).mean()["Sales"].to_frame().reset_index()
    sns.barplot(comp_year_sales[column], comp_year_sales["Sales"])
    plt.figure()


# Open/Closed -> high
# DayOfWeek -> medium
# Holiday -> medium
```
